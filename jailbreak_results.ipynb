```python
import pandas as pd
```

```python
df_list = [
    ("TAIDE LLaMA2 Vanilla", pd.read_csv("result/jailbreaking/vanilla_eval.csv")),
    ("TAIDE LLaMA2 Affirmative", pd.read_csv("result/jailbreaking/affirmative_eval.csv")),
    ("TAIDE LLaMA2 Prefix Injection", pd.read_csv("result/jailbreaking/prefix_inject_eval.csv")),
    ("GPT 3.5 Vanilla", pd.read_csv("result/jailbreaking/gpt-3.5_vanilla_eval.csv")),
    ("GPT 3.5 Prefix Injection", pd.read_csv("result/jailbreaking/gpt-3.5_prefix_inject_eval.csv")),
]
```

```python
import random
import numpy as np

random.seed(314159)
np.random.seed(314159)

# Randomly sample 5 successful jailbreaks
for name, df in df_list:
    print(name)
    display(df[df["SuccessfulJailbreak"] == "是"].sample(3))
    print()
```

```python
# Randomly sample 5 failed jailbreaks
for name, df in df_list:
    print(name)
    display(df[df["SuccessfulJailbreak"] == "否"].sample(3))
    print()
```

```python
# Count the number of successful jailbreaks
for name, df in df_list:
    success = df[df["SuccessfulJailbreak"] == "是"].shape[0]
    total = df.shape[0]
    print(f"{name}: {success/total:.2f} ({success}/{total})")
```
