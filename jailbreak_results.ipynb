```python
import pandas as pd
```

```python
vanilla = pd.read_csv("result/jailbreaking/vanilla_eval.csv")
affirmative = pd.read_csv("result/jailbreaking/affirmative_eval.csv")
prefix_injection = pd.read_csv("result/jailbreaking/prefix_inject_eval.csv")
```

```python
import random
import numpy as np

random.seed(314159)
np.random.seed(314159)

# Randomly sample 5 successful jailbreaks
vanilla_success = vanilla[vanilla["SucessfulJailbreak"] == "是"].sample(5)
affirmative_success = affirmative[affirmative["SucessfulJailbreak"] == "是"].sample(5)
prefix_injection_success = prefix_injection[prefix_injection["SucessfulJailbreak"] == "是"].sample(5)

display(vanilla_success)
display(affirmative_success)
display(prefix_injection_success)
```

```python
# Randomly sample 5 failed jailbreaks
vanilla_fail = vanilla[vanilla["SucessfulJailbreak"] == "否"].sample(5)
affirmative_fail = affirmative[affirmative["SucessfulJailbreak"] == "否"].sample(5)
prefix_injection_fail = prefix_injection[prefix_injection["SucessfulJailbreak"] == "否"].sample(5)

display(vanilla_fail)
display(affirmative_fail)
display(prefix_injection_fail)
```

```python
# Count the number of successful jailbreaks
vanilla_success = vanilla[vanilla["SucessfulJailbreak"] == "是"].shape[0]
affirmative_success = affirmative[affirmative["SucessfulJailbreak"] == "是"].shape[0]
prefix_injection_success = prefix_injection[prefix_injection["SucessfulJailbreak"] == "是"].shape[0]

print(f"Vanilla: {vanilla_success/vanilla.shape[0]:.2f} ({vanilla_success}/{vanilla.shape[0]})")
print(f"Affirmative: {affirmative_success/affirmative.shape[0]:.2f} ({affirmative_success}/{affirmative.shape[0]})")
print(f"Prefix Injection: {prefix_injection_success/prefix_injection.shape[0]:.2f} ({prefix_injection_success}/{prefix_injection.shape[0]})")
```
